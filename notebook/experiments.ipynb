```python
import pandas as pd
from scipy import stats

# 1. Filter out the continuous features
continuous_cols = df.select_dtypes(include=["float64"]).columns

# 2. Initialize a list to store statistical results
test_results = []

# 3. Loop through each continuous feature and perform the test
for col in continuous_cols:
    # Split the feature data based on the categorical target 'Diagnosis'
    group_0 = df[df["Diagnosis"] == 0][col].dropna()
    group_1 = df[df["Diagnosis"] == 1][col].dropna()

    # Skip if one of the groups is empty
    if len(group_0) == 0 or len(group_1) == 0:
        continue

    # Perform Independent Two-Sample t-Test
    # equal_var=False performs Welch's t-test, which doesn't assume equal variances
    t_stat, p_val = stats.ttest_ind(group_0, group_1, equal_var=False)

    is_significant = "Yes" if p_val < 0.05 else "No"

    # Save metrics to our dictionary
    test_results.append(
        {
            "Feature": col,
            "t-statistic": round(t_stat, 4),
            "p-value": f"{p_val:.4e}" if p_val < 0.0001 else round(p_val, 4),
            "Statistically Significant? (α=0.05)": is_significant,
        }
    )

# 4. Convert results into a readable DataFrame and display
results_df = pd.DataFrame(test_results)
print("\n--- Hypothesis Testing Results ( Welch's t-test ) ---")
print(results_df.to_string(index=False))
```

The Null Hypothesis ($H_0$): The mean value of the feature is identical across both Diagnosis = 0 and Diagnosis = 1.

The Alternative Hypothesis ($H_1$): The mean value of the feature is significantly different between the two groups.

Interpretation: If the p-value is less than 0.05, we reject the null hypothesis ($H_0$). It formally proves that the difference you observed visually in the bottom row of your graphs is statistically backed by the numbers.

```python
import pandas as pd
from scipy import stats

continuous_cols = df.select_dtypes(include=["float64"]).columns
test_results = []

for col in continuous_cols:
    group_0 = df[df["Diagnosis"] == 0][col].dropna()
    group_1 = df[df["Diagnosis"] == 1][col].dropna()

    if len(group_0) < 3 or len(group_1) < 3:
        continue  # Shapiro-Wilk requires at least 3 samples

    # 1. Check Normality using Shapiro-Wilk Test (p > 0.05 means Normal)
    _, p_norm_0 = stats.shapiro(group_0)
    _, p_norm_1 = stats.shapiro(group_1)

    is_normal_0 = p_norm_0 > 0.05
    is_normal_1 = p_norm_1 > 0.05

    # 2. Dynamically choose the correct hypothesis test
    if is_normal_0 and is_normal_1:
        test_used = "Welch's t-test"
        stat, p_val = stats.ttest_ind(group_0, group_1, equal_var=False)
    else:
        test_used = "Mann-Whitney U"
        stat, p_val = stats.mannwhitneyu(
            group_0, group_1, alternative="two-sided"
        )

    is_significant = "Yes" if p_val < 0.05 else "No"

    test_results.append(
        {
            "Feature": col,
            "Distribution": "Normal" if (is_normal_0 and is_normal_1) else "Non-Normal",
            "Test Performed": test_used,
            "Statistic": round(stat, 4),
            "p-value": f"{p_val:.4e}" if p_val < 0.0001 else round(p_val, 4),
            "Significant? (α=0.05)": is_significant,
        }
    )

# Convert and display results
results_df = pd.DataFrame(test_results)
print("\n--- Dynamic Hypothesis Testing Results ---")
print(results_df.to_string(index=False))
```


---

### 1. The Powerhouse Predictors (Strong Statistical Significance)

The clinical assessment metrics exhibit incredibly strong statistical significance. Their p-values are effectively $0$, meaning the probability that these differences happened by pure random chance is zero.

* **FunctionalAssessment ($p = 8.9971 \times 10^{-65}$):** The most statistically significant feature in the entire dataset. The rank distribution between healthy and diagnosed groups is drastically shifted.
* **ADL ($p = 6.0536 \times 10^{-53}$):** Strongly distinct rank differences. Lower functional performance heavily correlates with a positive diagnosis.
* **MMSE ($p = 6.6238 \times 10^{-28}$):** Massive statistical significance proving cognitive scores differ fundamentally between diagnostic classes.

### 2. The Subtle / Marginal Predictors

Two lifestyle and lipid metrics managed to cross the significance boundary ($\alpha = 0.05$), though their signals are visually subtle due to the massive overlap in the uniform distributions.

* **SleepQuality ($p = 0.0094$):** Statistically significant. Even though the overall distribution looked uniform, the rank sum test detected a reliable, non-random shift in sleep scores between the groups.
* **CholesterolHDL ($p = 0.0486$):** Barely significant ($p$ is very close to $0.05$). There is a minor statistical difference in the rank tracking of "good" cholesterol between the categories, but it is a weak signal.

### 3. The Pure Statistical Noise (Not Significant)

For these 7 features, we **fail to reject the null hypothesis**. The test confirms that their distribution differences are completely random, providing zero predictive utility.

* **Lifestyle & Physicality:** `BMI` ($p = 0.2211$), `AlcoholConsumption` ($p = 0.6892$), `PhysicalActivity` ($p = 0.7857$), and `DietQuality` ($p = 0.6878$) have massive p-values.
* **Remaining Lipids:** `CholesterolTotal` ($p = 0.7884$), `CholesterolLDL` ($p = 0.1379$), and `CholesterolTriglycerides` ($p = 0.2849$) show no meaningful correlation to the diagnosis.

* **Uninformative Categoricals:** Demographics (`Gender`, `Ethnicity`, `EducationLevel`), underlying medical conditions (`Diabetes`, `Hypertension`, `CardiovascularDisease`, `Depression`), and even standard cognitive identifiers in this subset (`Confusion`, `Forgetfulness`, `Disorientation`) track linearly with the class imbalance baseline. They offer very little individual discriminative boundary.
* **Modeling Action:** When building your classification pipeline, your tree-based models (like XGBoost or Random Forest) will likely rank `MemoryComplaints` and `BehavioralProblems` at the top of the feature importance charts along with the continuous features (`MMSE`, `ADL`, `FunctionalAssessment`) identified earlier.



```python
import pandas as pd
from scipy import stats

# 1. Filter out all 'int64' columns except the high-cardinality/continuous ones
all_int_cols = df.select_dtypes(include=["int64"]).columns
exclude_cols = ["SystolicBP", "DiastolicBP", "Age", "Diagnosis"]
categorical_int_cols = [
    col for col in all_int_cols if col not in exclude_cols
]

chi2_results = []

# 2. Loop through each categorical feature and run the Chi-Square test
for col in categorical_int_cols:
    # Create the contingency table (cross-tabulation)
    contingency_table = pd.crosstab(df[col], df["Diagnosis"])

    # Perform the Chi-Square Test of Independence
    chi2_stat, p_val, dof, expected = stats.chi2_contingency(contingency_table)

    # Determine significance at alpha = 0.05
    is_significant = "Yes" if p_val < 0.05 else "No"

    chi2_results.append(
        {
            "Feature": col,
            "Chi2 Statistic": round(chi2_stat, 4),
            "p-value": f"{p_val:.4e}" if p_val < 0.0001 else round(p_val, 4),
            "Significant? (α=0.05)": is_significant,
        }
    )

chi2_df = pd.DataFrame(chi2_results).sort_values(
    by="Chi2 Statistic", ascending=False
)
print("\n--- Chi-Square Test of Independence Results ---")
print(chi2_df.to_string(index=False))
```

1. The Dominant Categorical FeaturesOnly two categorical features show a statistically significant relationship with the diagnosis at the standard $\alpha = 0.05$ threshold. Their extremely large Chi-Square values and near-zero p-values mean the relationship is mathematically indisputable.

MemoryComplaints ($\chi^2 = 200.62$, $p = 1.5266 \times 10^{-45}$): This is by far the strongest categorical predictor. The incredibly small p-value proves that whether a patient has memory complaints is fundamentally linked to their diagnostic outcome.

BehavioralProblems ($\chi^2 = 106.88$, $p = 4.7314 \times 10^{-25}$): Another exceptionally powerful predictor. The high Chi-Square value confirms a strong, non-random distribution shift between the two classes.
